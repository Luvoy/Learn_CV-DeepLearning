## 监督学习 之 线性回归 Linear Regression

> 这是一个回归问题

**引言**: 线性回归, 就是通过很多点得到一条直线,通过直线预测未知的点

y - label

x - attribute/character

解法: Ordinary Least Square 普通最小二乘法

我们需要用一条线来代表这些点, 也就是hypothesis 
$$h_{\theta }(x) = \theta _{0}+\theta _{1}x$$
使得Squared Error Function:
$$Cost = J(\theta_{0},\theta_{1})= \frac{1}{2m}  \sum_{i=1}^{m}(y_{i}-h_{\theta }(x_{i}))^2$$
最小



J(\theta)被称为loss function\cost function\target function, **但是,它代表的是假定值和真实值的差异, 也就是预测的准确程度**

如何求最小值:
- 解析法:求J关于\theta的偏导（不好用）
- 数值法:
    - 只需一阶导, 梯度下降法
    - 二阶导,牛顿法

### 梯度下降法
$$\theta_{j}=\theta_{j}-\alpha \frac{\partial }{\partial \theta_{j}}J(\theta) $$

分开写,也就是:

$$\theta_{0}=\theta_{0}-\alpha \frac{\partial }{\partial \theta_{0}}J(\theta_{0},\theta_{1}) $$

$$\theta_{1}=\theta_{1}-\alpha \frac{\partial }{\partial \theta_{1}}J(\theta_{1},\theta_{1}) $$

对于每一步, 需要分别计算:

$$\frac{\partial}{\partial \theta_{0}}J(\theta_{0})=\frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x_{i})-y_{i})$$

$$\frac{\partial}{\partial \theta_{1}}J(\theta_{1})=\frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x_{i})-y_{i})\cdot x_{i}$$

对于 碗状的 ,很好用

对于很多坑的函数,, 需要加一点动量来梯度下降,以防止达到局部最优解就停止了, 我们要的是全局最优解

### 多变量线性回归

房子属性有很多,地段\面积等等,分量x1...x100, 但是都是一次方线性的,a1*x1+ ... + a100*x100

对应一个房价 y

但是不同的属性 尺度不一样(单位大小)

导致J的投影(等高线)变成椭圆,梯度下降就很曲折(zig zag)

因此需要归一化,Feature Scaling, 每个属性减去均值再除以标准差

等高线变成圆,可以减少迭代,提高训练速度

## 监督学习 之 Logistics回归 逻辑斯蒂回归
> 这是个分类问题

举例：f(size)= price 但是可能不在0到1之间

g(f(size))->(0,1)

然后>0.5分类为A, 否则b类

Sigmoid Function或者叫Logistic Function：（**Hypothesis**）
$$h_{\theta}(\vec{x}) = g(f(\vec{x}))=g(\vec{\theta}^T \cdot \vec{x}))= \frac{1}{1+e^{-\vec{\theta} ^T \vec{x}}}$$
相当于$$P(y=A)=h_{\theta}(\vec{x})$$
$$P(y=B)=1-h_{\theta}(\vec{x})$$
不妨另A=1,B=0即
$$P(y)=h_{\theta}(\vec{x})^y \cdot (1-h_{\theta}(\vec{x}))^{1-y}$$
通过Maximum likelihood estimation最大似然估计:
$$L(\theta)=\prod_{i=1}^{m}P(y_{i})$$
取对数得到**Cost Funtion**:
$$J(\theta)=-\frac{1}{m} \sum_{i=1}^{m}[y \log (h_{\theta}(x_{i}))+(1-y)\log(1-h_{\theta}(x_{i}))]$$
也可以通过线性回归来求
$$J(\theta)= \frac{1}{2m}  \sum_{i=1}^{m}(y_{i}-h_{\theta }(x_{i}))^2$$
梯度下降:
$$\frac {\partial}{\partial \theta_{j}}J(\theta_{j})=\frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x_{i})-y_{i})\cdot x_{j}$$

## RANSAC random sample consensus

- 确定模型是一种不变的形式, 待求参数不知道,但是固定是那几个

误差容忍(阈值)q

所有符合模型的点叫内点,  所有内点叫内点集

百分之多少的样本都在内点集里,就ok了

不满足的时候,就再从样本里取数据,重新计算参数, 创建新的内点集
