## 监督学习 之 线性回归 Linear Regression

> 这是一个回归问题

**引言**: 线性回归, 就是通过很多点得到一条直线,通过直线预测未知的点

y - label

x - attribute/character

解法: Ordinary Least Square 普通最小二乘法

我们需要用一条线来代表这些点, 也就是hypothesis 
$$h_{\theta }(x) = \theta _{0}+\theta _{1}x$$
使得Squared Error Function:
$$Cost = J(\theta_{0},\theta_{1})= \frac{1}{2m}  \sum_{i=1}^{m}(y_{i}-h_{\theta }(x_{i}))^2$$
最小



J(\theta)被称为loss function\cost function\target function, **但是,它代表的是假定值和真实值的差异, 也就是预测的准确程度**

如何求最小值:
- 解析法:求J关于\theta的偏导（不好用）
- 数值法:
    - 只需一阶导, 梯度下降法
    - 二阶导,牛顿法

### 梯度下降法
$$\theta_{j}=\theta_{j}-\alpha \frac{\partial }{\partial \theta_{j}}J(\theta) $$

分开写,也就是:

$$\theta_{0}=\theta_{0}-\alpha \frac{\partial }{\partial \theta_{0}}J(\theta_{0},\theta_{1}) $$

$$\theta_{1}=\theta_{1}-\alpha \frac{\partial }{\partial \theta_{1}}J(\theta_{1},\theta_{1}) $$

对于每一步, 需要分别计算:

$$\frac{\partial}{\partial \theta_{0}}J(\theta_{0})=\frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x_{i})-y_{i})$$

$$\frac{\partial}{\partial \theta_{1}}J(\theta_{1})=\frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x_{i})-y_{i})\cdot x_{i}$$

对于 碗状的 ,很好用

对于很多坑的函数,, 需要加一点动量来梯度下降,以防止达到局部最优解就停止了, 我们要的是全局最优解

### 多变量线性回归

房子属性有很多,地段\面积等等,分量x1...x100, 但是都是一次方线性的,a1*x1+ ... + a100*x100

对应一个房价 y

但是不同的属性 尺度不一样(单位大小)

导致J的投影(等高线)变成椭圆,梯度下降就很曲折(zig zag)

因此需要归一化,Feature Scaling, 每个属性减去均值再除以标准差

等高线变成圆,可以减少迭代,提高训练速度

## 监督学习 之 Logistics回归 逻辑斯蒂回归
> 这是个分类问题

举例：f(size)= price 但是可能不在0到1之间

g(f(size))->(0,1)

然后>0.5分类为A, 否则b类

Sigmoid Function或者叫Logistic Function：（**Hypothesis**）
$$h_{\theta}(\vec{x}) = g(f(\vec{x}))=g(\vec{\theta}^T \cdot \vec{x}))= \frac{1}{1+e^{-\vec{\theta} ^T \vec{x}}}$$
相当于$$P(y=A)=h_{\theta}(\vec{x})$$
$$P(y=B)=1-h_{\theta}(\vec{x})$$
不妨另A=1,B=0即
$$P(y)=h_{\theta}(\vec{x})^y \cdot (1-h_{\theta}(\vec{x}))^{1-y}$$
通过Maximum likelihood estimation最大似然估计:
$$L(\theta)=\prod_{i=1}^{m}P(y_{i})$$
取对数得到**Cost Funtion**:
$$J(\theta)=-\frac{1}{m} \sum_{i=1}^{m}[y \log (h_{\theta}(x_{i}))+(1-y)\log(1-h_{\theta}(x_{i}))]$$
也可以通过线性回归来求
$$J(\theta)= \frac{1}{2m}  \sum_{i=1}^{m}(y_{i}-h_{\theta }(x_{i}))^2$$
梯度下降:
$$\frac {\partial}{\partial \theta_{j}}J(\theta_{j})=\frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x_{i})-y_{i})\cdot x_{j}$$

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import time
# 线性回归模型 预测
def inference(w,b,x):
    return w*x + b

# 损失函数 定义
def loss_functin(w, b, x_list, gt_y_list):
    avg_loss = 0
    for i in range(len(x_list)):
        avg_loss +=  0.5*(w*x_list[i]+b - gt_y_list[i])**2
    avg_loss /= len(x_list)
    return avg_loss
```

```python
# 捏造数据
def gen_sample_data():
    w = random.randint(0,10) + random.random()
    b = random.randint(0, 5) + random.random()
    
    num_sample = 100
    x_list = []
    y_list = []
    print(w,b)
    for i in range(num_sample):
        x = random.randint(0,100) * random.random()
        y = w * x + b + random.random() * random.randint(-1, 100)
        
        x_list.append(x)
        y_list.append(y)
        
    return x_list, y_list

x_list, y_list = gen_sample_data()

import matplotlib.pyplot as plt

plt.figure()
plt.scatter(x_list, y_list)
plt.show()
```

```python
#单一样本带来的梯度
def gradient(pred_y, gt_y, x):
    diff = pred_y - gt_y
    dw = diff * x
    db = diff
    return dw,db
```

```python
#全部样本（batchsize）为w,b带来的更新
def cal_step_gradient(batch_x_list, batch_gt_y_list, w, b ,lr):
    avg_dw, avg_db = 0, 0
    batch_size = len(batch_x_list)
    for i in range(batch_size):
        pred_y = inference(w, b, batch_x_list[i])
        dw, db = gradient(pred_y, batch_gt_y_list[i], batch_x_list[i])
        avg_dw += dw
        avg_db += db
    avg_dw /= batch_size
    avg_db /= batch_size
    w -= lr * avg_dw
    b -= lr * avg_db
    return w,b
```

```python
def train(x_list, gt_y_list, batch_size, lr, max_iter):
    w = 0
    b = 0
    num_samples = len(x_list)
    for i in range(max_iter):
        batch_idxs = np.random.choice(len(x_list), batch_size) #随机抽取batch_size个样本的索引值
        batch_x = [x_list[j] for j in batch_idxs]
        batch_y = [gt_y_list[j] for j in batch_idxs]
        w, b = cal_step_gradient(batch_x, batch_y, w, b, lr)
        print('w:{0},b:{1}'.format(w,b))
        print('loss is {}'.format(loss_functin(w,b,x_list,gt_y_list)))
        time.sleep(0.1)
        
    return w,b
```

```python
train(x_list, y_list, 100, 0.001, 1000)
```

## RANSAC random sample consensus
条件:
- 数据(有噪音, 非常影响模型精确度)
- 确定模型是一种不变的形式, 待求参数不知道,但是固定是那几个

误差容忍(阈值)q

所有符合模型的点叫内点,  所有内点叫内点集

百分之多少的样本都在内点集里,就ok了

不满足的时候,就再从样本里取数据,重新计算参数, 创建新的内点集

假设样本中有nin个好数据,nout个噪点

内点占比t=nin/(nin+nout)

选N个点,他们都是内点的概率P=t^N

至少有一个外点的概率P=1-t^N

迭代k次能得到最好模型的概率,也就是至少使用一次全部内点就能计算出模型的概率:

P=1-(1-t^N)^k

k= (log (1-p))/log(1-t^N)

```python
import numpy as np
import matplotlib.pyplot as plt
import random

SIZE = 50
ERROR = 100

x = np.linspace(0, 10, SIZE)
y = 3 * x + 10
random_x = [x[i] + random.uniform(-0.5, 0.5) for i in range(SIZE)]
random_y = [y[i] + random.uniform(-0.5, 0.5) for i in range(SIZE)]
```

```python
for i in range(ERROR):
    random_x.append(random.uniform(0, 20))
    random_y.append(random.uniform(10, 40))

RANDOM_X = np.array(random_x)
RANDOM_Y = np.array(random_y)
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(RANDOM_X, RANDOM_Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression(fit_intercept=True)
reg.fit(RANDOM_X.reshape(-1, 1), RANDOM_Y.reshape(-1, 1))
slope = reg.coef_
intercept = reg.intercept_
PREDICT_Y = reg.predict(RANDOM_X.reshape(-1, 1))
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(RANDOM_X, RANDOM_Y)
ax1.plot(RANDOM_X, PREDICT_Y, c='red')
plt.show()
```

```python
iterations = 100
tolerent_sigma = 1
thresh_size = 0.2

best_slope = -1
best_intercept = 0
pretotal = 0

plt.ion()
plt.figure()
for i in range(iterations):
    sample_index = random.sample(range(SIZE + ERROR), 2)
    x_1 = RANDOM_X[sample_index[0]]
    x_2 = RANDOM_X[sample_index[1]]
    y_1 = RANDOM_Y[sample_index[0]]
    y_2 = RANDOM_Y[sample_index[1]]

    # y = ax +b
    slope = (y_2 - y_1) / (x_2 - x_1)
    intercept = y_1 - slope * x_1

    # calculate inliers
    total_inliers = 0
    for index in range(SIZE + ERROR):
        PREDICT_Y = slope * RANDOM_X[index] + intercept
        if abs(PREDICT_Y - RANDOM_Y[index]) < tolerent_sigma:
            total_inliers += 1

    if total_inliers > pretotal:
        pretotal = total_inliers
        best_slope = slope
        best_intercept = intercept

    if total_inliers > (SIZE + ERROR) * thresh_size:
        break

    plt.title(f"RANSAC in Linear Regression: Iter {i + 1}, Inliers {pretotal}")
    plt.scatter(RANDOM_X, RANDOM_Y)
    Y = best_slope * RANDOM_X + best_intercept
    plt.plot(RANDOM_X, Y,'black')
    plt.pause(1)
    plt.clf()
```
